```python
#Model 1.1
#Baseline model without HAC
import pandas as pd
import numpy as np
import statsmodels.api as sm

df = pd.read_excel(r"/content/Data_for_Project.xlsx", sheet_name='variables')

df['IIP'] = df['IIP(Proxy for Income)']
df['Cash'] = df['Value( Rupees Crores )(Cash)']

# Keep as DataFrame (IMPORTANT)
X = sm.add_constant(df[['IIP', 'Cash']])
Y = df['Value( Rupees Crores )(UPI)']

model_1 = sm.OLS(Y, X).fit()
print(model_1.summary())

from statsmodels.stats.diagnostic import het_breuschpagan

residuals = model_1.resid
exog = model_1.model.exog

bp_test = het_breuschpagan(residuals, exog)

print("LM stat of Breusch Pagan test:", bp_test[0])
print("p-value of Breusch Pagan Test:", bp_test[1])

```

```python
#Model 1.2
#Baseline model with HAC
import pandas as pd
import numpy as np
import statsmodels.api as sm

df = pd.read_excel(r"/content/Data_for_Project.xlsx", sheet_name='variables')

df['IIP'] = df['IIP(Proxy for Income)']
df['Cash'] = df['Value( Rupees Crores )(Cash)']

# Keep as DataFrame (IMPORTANT)
X = sm.add_constant(df[['IIP', 'Cash']])
Y = df['Value( Rupees Crores )(UPI)']

model_1 = sm.OLS(Y, X).fit(cov_type='HAC', cov_kwds={'maxlags':1})
print(model_1.summary())

from statsmodels.stats.diagnostic import het_breuschpagan

residuals = model_1.resid
exog = model_1.model.exog

bp_test = het_breuschpagan(residuals, exog)

print("LM stat of Breusch Pagan test:", bp_test[0])
print("p-value of Breusch Pagan Test:", bp_test[1])
```

```python
#model 6
#Model with seasonal variation
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Re-load df and create log-transformed columns to ensure robustness and self-containment
df = pd.read_excel(r"/content/Data_for_Project.xlsx", sheet_name='variables')
df['dif_log_UPI'] = np.log(df['Value( Rupees Crores )(UPI)'])
df['dif_log_Cash'] = np.log(df['Value( Rupees Crores )(Cash)'])
df['dif_log_Card'] = np.log(df['Value( Rupees Crores )(Card)'])
df['dif_log_IIP'] = np.log(df['IIP(Proxy for Income)'])
df['Date'] = pd.to_datetime(df['Month/Year'], format='%b-%y')
df['month'] = df['Date'].dt.month

map_temp = {
    1 : 1,
    2 : 0,
    3 : 1,
    4 : 0,
    5 : 1,
    6 : 0,
    7 : 0,
    8 : 1,
    9 : 0,
    10 : 1,
    11 : 0,
    12 : 1,
    }

df["month_dummy"] = df["month"].map(map_temp)
month_dummies=df['month_dummy']

# Calculate differences for log-transformed variables and drop the first NaN row
df_diff = df[['dif_log_UPI', 'dif_log_Cash', 'dif_log_Card', 'dif_log_IIP']].diff().dropna()

# Correctly add 'month_dummy' as a new column to df_diff, aligning by index
df_diff['month_dummy'] = month_dummies.loc[df_diff.index]

# X = sm.add_constant(df_diff)
X = sm.add_constant(df_diff.drop(columns=['dif_log_UPI']))
Y = df_diff['dif_log_UPI']

model3 = sm.OLS(Y, X).fit(cov_type='HAC', cov_kwds={'maxlags':1})
print(model3.summary())
```

```python
#Model 7
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Load data
df = pd.read_excel(r"/content/Data_for_Project.xlsx", sheet_name='variables')

# Create log variables with corrected column names
df['dif_log_UPI'] = np.log(df['Value( Rupees Crores )(UPI)'])
df['dif_log_Cash'] = np.log(df['Value( Rupees Crores )(Cash)'])
df['dif_log_Card'] = np.log(df['Value( Rupees Crores )(Card)'])
df['dif_log_IIP'] = np.log(df['IIP(Proxy for Income)'])
# Ensure COVID dummy exists (0/1)
# df['COVID Dummy'] should already be in your data

# Create differenced data
df_diff = df[['dif_log_UPI', 'dif_log_Cash', 'dif_log_Card', 'dif_log_IIP']].diff().dropna()

# Align COVID dummy
df_diff['covid_dummy'] = df['COVID Dummy'].iloc[1:]

# Create interaction (IMPORTANT: use log_Card, not level)
df_diff['covid*card'] = df_diff['dif_log_Card'] * df_diff['covid_dummy']

# Define X and Y
X = sm.add_constant(df_diff[['dif_log_Cash', 'dif_log_Card', 'dif_log_IIP', 'covid_dummy', 'covid*card']])
Y = df_diff['dif_log_UPI']

# Run model
model_IT = sm.OLS(Y, X).fit(cov_type='HAC', cov_kwds={'maxlags':1})
print(model_IT.summary())
```
